```python
import pandas as pd
import re
import numpy as np

## Dataset er file path
file_path = "Data sources/SMSSpamCollection"

## Dataset load korar jonno 
def load_sms_dataset(file_path):
    try:
        df = pd.read_csv(file_path, sep="\t", header=None, names=["label", "message"], encoding="latin-1")
    except Exception:
        df = pd.read_csv(file_path, sep="\t", header=None, names=["label", "message"], encoding="utf-8", errors="replace")
    df = df.dropna(subset=["label", "message"]).reset_index(drop=True)
    return df

## Text clean korar jono 
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = text.strip()
    tokens = text.split()   
    return tokens

## Vocabulary toyri korar jonno function
def build_vocabulary(documents):
    vocab = set()
    for tokens in documents:
        vocab.update(tokens)
    return list(vocab)

## Bag of Words feature toyri korar jonno function
def bag_of_words(documents, vocabulary=None):
    
    if vocabulary is None:
        vocabulary = build_vocabulary(documents)
    
    vocabIndex = {word: i for i, word in enumerate(vocabulary)}
    
    matrix = []
    
    for tokens in documents:
        vector = [0] * len(vocabulary)
        
        for word in tokens:
            if word in vocabIndex:
                vector[vocabIndex[word]] += 1
        
        matrix.append(vector)
    
    return np.array(matrix), vocabulary

## Tf-idf feature toyri korar jonno function
def tf_idf(documents, vocabulary=None, document_frequency=None, total_documents=None):
    
    if vocabulary is None:
        vocabulary = build_vocabulary(documents)
    
    vocabIndex = {word: i for i, word in enumerate(vocabulary)}
    N = len(documents)
    
    #Term Frequency (TF)
    tfMatrix = np.zeros((N, len(vocabulary)))
    
    for i, tokens in enumerate(documents):
        for word in tokens:
            if word in vocabIndex:
                tfMatrix[i][vocabIndex[word]] += 1
    
    #Document Frequency (DF)
    df = np.zeros(len(vocabulary))
    
    for i, word in enumerate(vocabulary):
        for tokens in documents:
            if word in tokens:
                df[i] += 1
    
    # IDF
    idf = np.log((N + 1) / (df + 1)) + 1
    
    # TF-IDF
    tfidfMatrix = tfMatrix * idf
    
    return tfidfMatrix, vocabulary

## Naive Bayes train korar jonno function
def train_naive_bayes(x_train_features, x_test_features, y_train):
    # Ekhane Naive Bayes model train hobe
    pass

## Model evaluate korar jonno function
def evaluate_model(y_test, predictions):
    # Ekhane accuracy o other metric check hobe
    pass

## Main flow er structure
# 1. Dataset load korbe
# 2. Text clean korbe
# 3. Train-test split korbe
# 4. BoW and tf-idf features banabe
# 5. Naive Bayes train korbe
# 6. Result evaluate korbe
```
